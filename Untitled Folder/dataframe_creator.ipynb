```python
import sys
import numpy as np
import pandas as pd

# how the plots in the notebook
%matplotlib notebook

## CONSTANTS
LABEL_TIME_IDX = 3

imu_file = "/Volumes/CAR-PROJECTS/Martelaro_IsNowAGoodTime/ParticipantRuns/2017-08-11_m028/SYNCED-DATA/imuData_out.csv"
labels_file = "/Volumes/CAR-PROJECTS/Martelaro_IsNowAGoodTime/ParticipantRuns/2017-08-11_m028/SYNCED-DATA/labels.csv"

label_list = []
# create a vector of the label times
with open(labels_file, "r") as labels:
    for line in labels:
        data = line.strip().split(",")
        if data[3] != "Annotation":
            label_list.append(data[3])
label_list
```

```python
imu_df = pd.read_csv(imu_file)
imu_df
```

```python
imu_df['accelX_smth'] = imu_df['accelX'].rolling(window=10).mean()
imu_df.plot('time',y = ['accelX_smth','accelY'])
imu_df['accelX'].corr(imu_df['accelY'])
```

```python
d = []
# define how many seconds around the label we want to sample
fwd_window = 5
rev_window = 5

for time in label_list:
    # get the data between the windows values around the label time
    selection = imu_df[imu_df['time'].between(float(time) - rev_window, float(time) + fwd_window)]
    accelX = selection['accelX'].mean()
    gyroZ = selection['gyroZ'].mean()
    d.append({'time':float(time), 'gyroZ':gyroZ, 'accelX':accelX})
```

```python
output_df = pd.DataFrame(d)
output_df
```
